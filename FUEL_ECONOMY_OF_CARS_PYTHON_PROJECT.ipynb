```python
'''Also you need to merge all the datasets with same columns names by adding a column of year to
differentiate the data for different year (drop the extra columns. Also make sure that data types of
merged columns are same.
Find the appropriate number of clusters from the data and conclude the properties of the clusters.
Create a Predictive model to predict whether a vehicle is SmartWay vehicle or not using the
Classification Algorithm (Logistic Regression and Decision Tree). Also conclude, which is the best
model and why not other.
Also, create a model to predict the City Mileage per gallon (City MPG) using the best features'''
```

```python
# Add 'Year' column to each dataset
df_08['Year'] = 2008
df_09['Year'] = 2009
df_10['Year'] = 2010
df_11['Year'] = 2011
df_12['Year'] = 2012
df_13['Year'] = 2013
df_14['Year'] = 2014
df_15['Year'] = 2015
df_16['Year'] = 2016
df_17['Year'] = 2017
df_18['Year'] = 2018
```

```python
# Merge datasets
df_list = [df_08, df_09,df_10,df_11,df_12,df_13,df_14,df_15,df_16,df_17,df_18]  # Add all dataframes to this list
df_merged = pd.concat(df_list, ignore_index=True)
```

```python
# Drop extra columns and ensure data types are consistent
df_merged = df_merged.loc[:, ~df_merged.columns.duplicated()]
df_merged = df_merged.convert_dtypes()
```

```python
# Check the merged dataframe
print(df_merged.info())
```

```python
df_merged
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Select features for clustering
features = df_merged[['city_mpg', 'hwy_mpg', 'cmb_mpg']]  # Select relevant features

# Determine the optimal number of clusters using the elbow method
sse = {}
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(features)
    sse[k] = kmeans.inertia_

# Plot the elbow curve
plt.figure(figsize=(10, 6))
sns.lineplot(x=list(sse.keys()), y=list(sse.values()), marker='o')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score

# Prepare the data
df_merged['smartway'] = df_merged['smartway'].apply(lambda x: 1 if x == 'yes' else 0)
X = df_merged[['city_mpg', 'hwy_mpg', 'cmb_mpg']]  # Select relevant features
y = df_merged['smartway']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Logistic Regression
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
y_pred_log_reg = log_reg.predict(X_test)
print("Logistic Regression:")
print(classification_report(y_test, y_pred_log_reg))

# Decision Tree
decision_tree = DecisionTreeClassifier(random_state=42)
decision_tree.fit(X_train, y_train)
y_pred_decision_tree = decision_tree.predict(X_test)
print("Decision Tree:")
print(classification_report(y_test, y_pred_decision_tree))

# Compare Models
accuracy_log_reg = accuracy_score(y_test, y_pred_log_reg)
accuracy_decision_tree = accuracy_score(y_test, y_pred_decision_tree)

print(f"Logistic Regression Accuracy: {accuracy_log_reg}")
print(f"Decision Tree Accuracy: {accuracy_decision_tree}")

if accuracy_log_reg > accuracy_decision_tree:
    print("Logistic Regression is the better model.")
else:
    print("Decision Tree is the better model.")
```

```python
from sklearn.ensemble import RandomForestRegressor

# Prepare the data for City MPG prediction
features = df_merged[['cyl', 'trans', 'drive', 'Year']]
target = df_merged['city_mpg']

# Convert categorical features to numeric using one-hot encoding
features = pd.get_dummies(features)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)

# Random Forest Regressor
rf_regressor = RandomForestRegressor(random_state=42)
rf_regressor.fit(X_train, y_train)
y_pred_rf = rf_regressor.predict(X_test)

# Evaluate the model
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred_rf)
r2 = r2_score(y_test, y_pred_rf)

print(f"Random Forest Regressor MSE: {mse}")
print(f"Random Forest Regressor R²: {r2}")
```
